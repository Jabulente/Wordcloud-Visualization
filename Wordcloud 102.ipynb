```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import string
from wordcloud import STOPWORDS
from nltk.stem import WordNetLemmatizer
```

### Text Cleansing and Preprocessing

- Converts the text data to lowercase to maintain uniformity.
- Removes punctuation using Python's built-in `string.punctuation`.
- Splits the text into a list of individual words (tokens).
- Filters out common stopwords that don’t contribute much meaning.
- Optionally reduces words to their base or root form using a lemmatizer.
- Reassembles the processed words into a single string to be used for wordcloud generation.

```python
def preprocess_text(file_path):
    with open(file_path, 'r') as file:
        text_data = file.read()

    text_data = text_data.lower()
    text_data = text_data.translate(str.maketrans('', '', string.punctuation))
    words = text_data.split() #  Tokenization
    filtered_words = [word for word in words if word not in STOPWORDS] # 5. Remove Stopwords
    
    #lemmatizer = WordNetLemmatizer()
    #lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words] # 6. Lemmatization (Optional)
    
    processed_text = ' '.join(filtered_words)
    return processed_text
```

<h1 style='font-size: 45px; color: crimson; font-family: Dubai; font-weight: 600'>LOVE 101</h1>

```python
file_path = 'Datasets/102.txt'
processed_text = preprocess_text(file_path)
print(processed_text[:2000])
```

```python
wordcloud = WordCloud(width=1400, height=700, background_color='black').generate(processed_text)
plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import string
from nltk.corpus import stopwords

stopwords_kiswahili = [
    "na", "kwa", "au", "na", "al", "katika", "hiyo", "yeye", "wao", "hii", 
    "kwamba", "kuhusu", "kama", "za", "hiyo", "kwa hivyo", "kisha", "wa", 
    "hii", "tu", "kwa hiyo", "jinsi", "wale", "nyingi", "zaidi", "sisi", 
    "kusema", "kila", "moja", "watu", "lakini", "vile", "pia", "aliyo", 
    "yote", "hata", "aina", "kutafuta", "pamoja", "ya", "kuwa"
]


def remove_stopwords(text):
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords_kiswahili]
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text

cleaned_text = remove_stopwords(processed_text)

print("Cleaned Text: ", cleaned_text)
```

```python
wordcloud = WordCloud(width=1400, height=700, background_color='black').generate(cleaned_text)
plt.figure(figsize=(14, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
